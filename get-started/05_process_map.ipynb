```python
log = pm4py.read_xes('../data/running-example.xes')
log
```

```python
dfg, start_activities, end_activities = pm4py.discover_dfg(log)
dfg, start_activities, end_activities
```

```python
pm4py.view_dfg(dfg, start_activities, end_activities)
```

A process map contains activities and connections (by means of arcs) between them. A connection between two activities usually means that there some form of precedence relation. In its simplest form, it means that the ‘source’ activity directly precedes the ‘target’ activity

The above graph is a ‘Directly Follows Graph’ (DFG)-based process map
